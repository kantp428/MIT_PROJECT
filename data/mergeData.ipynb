```python
import pandas as pd
from pathlib import Path

# โฟลเดอร์ข้อมูล
data_path = Path("processed_data")

# รวมไฟล์ที่ต้องการ
files = [
    "daily_averageDN.csv",
    "daily_averageBN.csv",
    "daily_averageKPS.csv"
]

# อ่านและเก็บใน list
df_list = []
for file in files:
    df = pd.read_csv(data_path / file)
    df_list.append(df)

# รวมข้อมูล
all_data = pd.concat(df_list, ignore_index=True)

# แปลง date เป็น datetime (สำคัญสำหรับ time-series)
all_data['date'] = pd.to_datetime(all_data['date'])

# เรียงตามวัน
all_data = all_data.sort_values('date').reset_index(drop=True)

# บันทึกไฟล์
output_path = data_path / "daily_average_all.csv"
all_data.to_csv(output_path, index=False)
all_data = all_data.drop(columns=['Unnamed: 0', 'index'], errors='ignore')

print("✅ Merge complete:", output_path)
```
