```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pm = 'rawdata/dn_pm.csv'
weather = 'rawdata/dn_weather.csv'
location = 'DN'
pm_data = pd.read_csv(pm)
weather_data = pd.read_csv(weather)
```

```python
pm_data.head()
```

```python
weather_data.head()
```

```python
pm_data['time'] = pd.to_datetime(pm_data['time'])
daily_avg = (
    pm_data
    .set_index('time')  
    .resample('D')       
    .mean()        
    .reset_index()    
)
daily_avg = daily_avg.rename(columns={
    'time': 'date',
    'us_aqi_pm2_5 (USAQI)': 'pm'
})
daily_avg["location"] = location
daily_avg = daily_avg.reset_index()
daily_avg.head()
```

```python
weather_data['date']= pd.to_datetime(weather_data['date'])
weather_data = weather_data.rename(columns={
    'temperature_2m_mean': 'temp',
    'dew_point_2m_mean': 'dew_point',
    'relative_humidity_2m_mean': 'humidity',
    'pressure_msl_mean': 'pressure',
    'wind_speed_10m_mean': 'wind_speed',
    'winddirection_10m_dominant': 'wind_direction',
    'precipitation_sum': 'precipitation'
})
weather_data.head()
```

```python
print(daily_avg.shape, weather_data.shape)
```

```python
merged_df = pd.merge(
    daily_avg,
    weather_data,
    on='date',          # key ที่ใช้เชื่อม
    how='inner'         # เอาเฉพาะวันที่ตรงกัน
)

merged_df.head()
```

```python
print(merged_df.shape)
```

```python
merged_df.to_csv('processed_data/daily_average'+location+'.csv')
print("Data saved successfully!")
```
